```python
import threading
import pandas as pd
import numpy as np
from datetime import datetime
import glob
import random
import os
import ntpath
import warnings
warnings.filterwarnings('ignore')
```

```python
def procesar(csv,jump,regressors):
    df=pd.read_csv(csv)
    df = df.dropna()
    df['Tiempo']=df['Fecha'] + ' ' + df['Hora']
    df['Tiempo'] = pd.to_datetime(df.Tiempo)
    df['Dia']=df['Tiempo'].dt.strftime('%j').astype(int)
    df['Minuto']=df['Tiempo'].dt.hour*60+df['Tiempo'].dt.minute
    data_pack=[]
    first_row=0
    length=len(df)
    for i in range(length-1):
        if (df['Tiempo'][i+1]-df['Tiempo'][i]!=pd.Timedelta('0 days 00:15:00') or (i==length-1)):
            data_pack.append(df[first_row:i+1])
            first_row=i+1
    data_pack2=[]
    for data in data_pack:
        for i in range(regressors):
            new_df=data[data.index%jump==i]
            new_df=new_df.drop(['Fecha', 'Hora','Codigo','Tiempo'], axis=1)
            if (len(new_df)>regressors):
                data_pack2.append(new_df)
    return data_pack2    
```

```python
def copiarNpy(f,jump,regressors,outpath,number):
    numberstr=str(number)
    print('Inicia Thread '+numberstr)
    train=0.6
    test=0.2    
    
    final=f.replace("cleared",outpath, 1)
    outdir, outname = ntpath.split(final)

    fullname = os.path.join(outdir, os.path.splitext(outname)[0])

    dflist=procesar(f,jump,regressors)
    i=1
    print(f)
    for df in dflist:
        array=df.to_numpy(dtype=float)
        number=random.random()
        if (number<train):
            finalname=fullname.replace("conjunto","train", 1)
        elif (number>1-test):
            finalname=fullname.replace("conjunto","test", 1)
        else:
            finalname=fullname.replace("conjunto","val", 1)
        np.save(finalname+'_'+str(i), array)
        #print('Guarda Thread '+numberstr)
        i+=1
    print('Termina Thread '+numberstr)
```

```python
def csvToNpy(jump,regressors):
    path = '.\\cleared\\'    
    outpath='npy_jump_reg'+str(jump)+'_'+str(regressors)+'\\conjunto'
    
    files = [f for f in glob.glob(path + "**/*.csv", recursive=True)]
    
    for f in files:
        final=f.replace("cleared",outpath, 1)
        outdir, outname = ntpath.split(final)
        for folder in ["train","val","test"]:
            direc=outdir.replace("conjunto",folder, 1)
            if not os.path.exists(direc):
                os.makedirs(direc)        

    i=1
    for f in files:
        hilo = threading.Thread(target=copiarNpy, args=(f,jump,regressors,outpath,i,))
        hilo.start()
        i+=1

    print("listo")
    pass
```
